```python
def inspect_df(df, name="DataFrame"):
    print(f"==== {name} ====")
    if df is None:
        print("Result is None")
        return
    if getattr(df, "empty", True):
        print("DataFrame is empty")
        return
    
    print("shape:", df.shape)
    print("columns:", list(df.columns))
    print("\ndtypes:")
    print(df.dtypes)
    print("\nhead:")
    display(df.head())
    print("\ntail:")
    display(df.tail())
    print("\nmissing values:")
    display(df.isna().sum())
```

```python
def clean_price_df(df, code_col_name):
    tmp = df.copy()
    
    tmp["日期"] = pd.to_datetime(tmp["日期"])
    
    numeric_cols = ["开盘", "收盘", "最高", "最低", "成交量", "成交额", "涨跌幅", "涨跌额", "换手率"]
    for col in numeric_cols:
        if col in tmp.columns:
            tmp[col] = pd.to_numeric(tmp[col], errors="coerce")
    
    tmp = tmp.sort_values("日期").reset_index(drop=True)
    return tmp
```

```python
stock_histories_clean = {}

for code, df in stock_histories_long.items():
    stock_histories_clean[code] = clean_price_df(df, "股票代码")
    print(code, selected_stocks[code], stock_histories_clean[code].shape)
```

```python
def add_moving_averages(df):
    tmp = df.copy()
    tmp["MA20"] = tmp["收盘"].rolling(20).mean()
    tmp["MA60"] = tmp["收盘"].rolling(60).mean()
    tmp["MA120"] = tmp["收盘"].rolling(120).mean()
    tmp["MA250"] = tmp["收盘"].rolling(250).mean()
    return tmp
```

```python
stock_ma = {}

for code, df in stock_histories_clean.items():
    stock_ma[code] = add_moving_averages(df)
    print("=" * 60)
    print(code, selected_stocks[code])
    display(stock_ma[code][["日期", "收盘", "MA20", "MA60", "MA120", "MA250"]].tail())
```

```python
index_ma = {}

for code, df in index_histories_clean.items():
    index_ma[code] = add_moving_averages(df)
    print("=" * 60)
    print(code, index_ids[code])
    display(index_ma[code][["日期", "收盘", "MA20", "MA60", "MA120", "MA250"]].tail())
```

```python
def add_returns(df):
    tmp = df.copy()
    tmp["return_20d"] = tmp["收盘"] / tmp["收盘"].shift(20) - 1
    tmp["return_60d"] = tmp["收盘"] / tmp["收盘"].shift(60) - 1
    tmp["return_120d"] = tmp["收盘"] / tmp["收盘"].shift(120) - 1
    return tmp
```

```python
stock_features = {}

for code, df in stock_ma.items():
    stock_features[code] = add_returns(df)
    print("=" * 60)
    print(code, selected_stocks[code])
    display(
        stock_features[code][
            ["日期", "收盘", "return_20d", "return_60d", "return_120d"]
        ].tail()
    )
```

```python
index_features = {}

for code, df in index_ma.items():
    index_features[code] = add_returns(df)
    print("=" * 60)
    print(code, index_ids[code])
    display(
        index_features[code][
            ["日期", "收盘", "return_20d", "return_60d", "return_120d"]
        ].tail()
    )
```

```python
sse_df = index_features["000001"][["日期", "收盘"]].rename(columns={"收盘": "上证收盘"})

merged_vs_sse = {}

for code, df in stock_features.items():
    merged = df[["日期", "收盘"]].merge(sse_df, on="日期", how="inner")
    merged["relative_strength_vs_sse"] = merged["收盘"] / merged["上证收盘"]
    merged_vs_sse[code] = merged
    
    print("=" * 60)
    print(code, selected_stocks[code])
    print("merged rows:", len(merged))
    display(merged.tail())
```

```python
csi300_df = index_features["000300"][["日期", "收盘"]].rename(columns={"收盘": "沪深300收盘"})

merged_vs_csi300 = {}

for code, df in stock_features.items():
    merged = df[["日期", "收盘"]].merge(csi300_df, on="日期", how="inner")
    merged["relative_strength_vs_csi300"] = merged["收盘"] / merged["沪深300收盘"]
    merged_vs_csi300[code] = merged
    
    print("=" * 60)
    print(code, selected_stocks[code])
    print("merged rows:", len(merged))
    display(merged.tail())
```
